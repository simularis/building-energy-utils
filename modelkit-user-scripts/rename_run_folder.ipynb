```python
import os

# keys are original names, values are renamed names
rename_dict = {
    "Inst-0.81HI":"RG-WaterHtg_eq-Instant_UEF-0.81 UEF-HI",
    "Inst-0.87HI":"RG-WaterHtg_eq-Instant_UEF-0.87 UEF-HI",
    "Inst-0.92HI":"RG-WaterHtg_eq-Instant_UEF-0.92 UEF-HI",
    "Inst-0.96HI":"RG-WaterHtg_eq-Instant_UEF-0.96 UEF-HI",
    "Strg40-0.64HI":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.64 UEF-HI",
    "Strg40-0.60EF":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.60 EF",
    "Strg40-0.58MD":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.58 UEF-MD",
    "Inst-0.81MD":"RG-WaterHtg_eq-Instant_UEF-0.81 UEF-MD",
    "Inst-0.87MD":"RG-WaterHtg_eq-Instant_UEF-0.87 UEF-MD"
}    
    
    
    
#    "Strg30-0.78HI":"RG-WaterHtg_eq-Stor_UEF-30 gal-0.78 UEF-HI",
#    "Strg40-0.58MD":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.58 UEF-MD",
#    "Strg40-0.60EF":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.60 EF",
#    "Strg40-0.64MD":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.64 UEF-MD",
#    "Strg40-0.68HI":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.68 UEF-HI",
#    "Strg40-0.78HI":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.78 UEF-HI",
#    "Strg50-0.56MD":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.56 UEF-MD",
#    "Strg50-0.59EF":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.59 EF",
#    "Strg50-0.64MD":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.64 UEF-MD",
#    "Strg50-0.63HI":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.63 UEF-HI",
#    "Strg50-0.68HI":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.68 UEF-HI",
#    "Strg50-0.78HI":"RG-WaterHtg_eq-Stor_UEF-50 gal-0.78 UEF-HI",
#    "Strg40-0.64HI":"RG-WaterHtg_eq-Stor_UEF-40 gal-0.64 UEF-HI"  



def rename_folders(root_folder, rename_dict):
    # Traverse the directory structure
    for dirpath, dirnames, filenames in os.walk(root_folder, topdown=False):
        for dirname in dirnames:
            if dirname in rename_dict:
                old_path = os.path.join(dirpath, dirname)
                new_path = os.path.join(dirpath, rename_dict[dirname])
                print(f'Renaming "{old_path}" to "{new_path}"')
                os.rename(old_path, new_path)

# Define the root folder - can be the folder directory before run folder
root_folder = r"E:\SWWH013\residential measures\SWWH013-04 Tankless Water Heater\SWWH013-04 Tankless Water Heater_SFm_New"

# Call the function to rename folders
rename_folders(root_folder, rename_dict)

```

```python
def check_folders(root_folder, rename_dict):
    original_names = set(rename_dict.keys())
    found_folders = []

    # Traverse the directory structure
    for dirpath, dirnames, filenames in os.walk(root_folder):
        for dirname in dirnames:
            if dirname in original_names:
                found_folders.append(os.path.join(dirpath, dirname))

    if found_folders:
        print("Folders with original names found:")
        for folder in found_folders:
            print(folder)
    else:
        print("No folders with original names found. All folder names have been changed.")

# Call the function to check for folders with original names
check_folders(root_folder, rename_dict)
```
