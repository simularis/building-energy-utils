```python
import pandas
from pathlib import Path
import os
import datetime
```

```python
root = Path(r"E:/SWHC046/commercial measures/SWHC046-04 Pkg HP AC Com")
#root = Path(r"E:/SWHC046/commercial measures/SWHC046-04 Pkg HP AC Com/SWHC046-04 Pkg HP AC Com_Ex")
```

```python
root.lstat()
```

```python
list(root.glob('*'))
```

```python
results = []
for f in root.glob('*'):
    result = [
        root.as_posix(),
        f.name,
        f.is_file(),
        f.is_dir(),
        datetime.datetime.fromtimestamp(f.lstat().st_mtime),
        len(list(f.glob('**/instance-out.sql'))),
    ]
    results.append(result)
df_file_info = pandas.DataFrame(results, columns=['parent','filename','is_file','is_dir','mtime','sim instances'])
```

```python
df_file_info
```

```python
df_file_info.to_csv(root / 'folder_info.csv', index=False)
```
